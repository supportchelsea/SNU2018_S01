#### # by 함수
데이터 위치를 참조할 수 있는 데이터 구조가 만들어짐

by(데이터 = 벡터, 인다이스 = 인덱스!! 데이터와 똑같은 길이의 벡터가 들어가야함,
 함수= 원하는 기능을 쓰고, na.rm=결측치 제거)
 
단점
: 나누고 싶은 범주 많을때 거기에 해당하는 index array를 생성해줘야함

```r
by( data = surveys$weight, INDICES =  surveys$sex, 
    FUN  = mean, na.rm = TRUE)
```

```r
a = by( data = surveys$weight, INDICES =  surveys$sex, 
    FUN  = mean, na.rm = TRUE)
class(a)
a= unclass(a)
str(a)
unlist(a)
```

#### # aggregate 함수
    fomula를 이용
    회귀모형과 비슷
    
    복수 개의 범주로 구분가능 -> 인덱스 자동생성

```r
aggregate(formula = weight ~ sex, data = surveys,
          FUN = mean, na.rm = TRUE)
```

```r
aggregate(formula = weight ~ sex + species_id, 
          data = surveys, FUN = mean, na.rm = TRUE)

# 레벨별로 여러가지 통계량을 한번에 알고 싶을때는 적합하지 않음
```

```r
# melf : wide -> long

if (!require(dplyr)) { install.packages("dplyr") ; library(dplyr) }
if (!require(reshape2)) { install.packages("reshape2") ; library(reshape2) }

head(airquality, n = 3) #wide data

head(melt(airquality), n = 3) #long data 

names(airquality) <-  tolower(names(airquality)) #모든 문자를 소문자로 버꿔라
melt(data = airquality) %>% head(n=3) 
#아무런 조건없이 데이터를 멜트시키면, 순서대로 풀림
# 현재 값이 오존이라는 변수에 값이 41이다
```

```r
# melt의 식별자 지정 : 어떠한 단위로 melt를 해줄 것인가??
# id.vars을 지정해줌으로써 long format은 행을 구분해 줄 수 있는 식별자
names(airquality) <-  tolower(names(airquality))
aql <- melt(data = airquality, id.vars= c("month","day")) # var별로 멜트를 하여라
head(aql, n = 3)
```
