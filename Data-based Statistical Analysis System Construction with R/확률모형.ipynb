## 행렬

```r
set.seed(1)
X = matrix(runif(15),5,3)
```

```r
X[3,2]
X[,2,drop=FALSE] #매트릭스 구조 그대로 유지
X[4,,drop=FALSE] #drop안하면 벡터형태로 반환
X[,2]
```

```r
# 전치행렬
X
Xt = t(X)
Xt

X[,1]
Xt[1,] # 서로같음을 알 수 있음
```

```r
#정방행렬 : 정사각형 행렬
Xt%*%X
```

```r
# 역행렬
A = Xt%*%X
B = solve(A)

A%*%B

B%*%A
```

```r
# R연습

a = matrix(runif(3),3,1)
at = t(a)

at%*%a

tX = t(X)
tX%*%X

solve(tX%*%X)
at%*%solve(tX%*%X)%*%a

```

```r
# 다변량 정규분포의 pdf계산
#R연습

X = matrix(c(1,0,1/2),3,1)
Mu = matrix(c(0,1,-1),3,1)
cov = matrix(c(1,0.5,0,0.5,1,0.3,0,0.3,1),3,3,byrow = T)
s = -1/2*(t(X-Mu)%*%solve(cov)%*%(X-Mu))
exp(s)
```

```r
# t-test
n = 1000
testResult <- rep(0,n)
for (i in 1:n)
{
  x <- rep(0,20)
  for (j in 2:20) x[j] <- 0.9*x[j-1] + rnorm(1)
  testfit <- t.test(x,mu=0)
  testResult[i] <- testfit$p.value
}
mean(testResult>0.05)

testfit
```
